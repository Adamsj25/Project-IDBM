```python
import requests
import re

```

```python
url = 'https://www.metacritic.com/browse/movie/all/all/2008/metascore/?page=1'
headers = {'User-Agent':'Mozilla/5.0'}
response = requests.get(url,headers=headers)
html_content = response.text
```

```python
#extract the information from title, description, release_dates, metascores,and url(replace with the raw url)
titles = re.findall(r'<div data-title="(.*?)"', html_content)
description = re.findall(r'<div class="c-finderProductCard_description"><span>(.*?)\s?<\/span>',html_content)
release_dates = re.findall(r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s+\d{1,2},\s+\d{4}\b', html_content)
metascores = re.findall(r'div title="Metascore (.*?)"',html_content)
movies_urls_raw = re.findall(r'<img src=(.*).height="132" width="88" style="display:none;">',html_content)
movies_urls = [url.replace('&amp;','&').replace(';', '') for url in movies_urls_raw]
```

```python
#combined the extracted data into an array of tuples
movies_data = list(zip(titles,description,release_dates,metascores,movies_urls))
```

```python
#print out array into a nice formatted way
for movie in movies_data:
    print(f"titles: {movie[0]}")
    print(f"description:{movie[1]}")
    print(f"release_dates:{movie[2]}")
    print(f"metascores:{movie[3]}")
    print(f"url:{movie[4]}")
    print("---------------------------------")
```
